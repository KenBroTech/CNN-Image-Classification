```python
import keras
import cv2
import sklearn
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator as IDG
from tensorflow.keras.preprocessing import image
```

```python
import cv2
import time

TIMER = int(3)
webcam = cv2.VideoCapture(0)
while True:
    ret, img = webcam.read()
    cv2.imshow('a', img)
    k = cv2.waitKey(125)
    if k == ord('q'):
        prev = time.time()
        while TIMER >= 0:
            ret, img = webcam.read()
            font = cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
            cv2.putText(img, str(TIMER),(200, 250), font, 7, (0, 255, 255), 4, cv2.LINE_AA)
            cv2.imshow('a', img)
            cv2.waitKey(125)
            cur = time.time()
            if cur-prev >= 1:
                prev = cur
                TIMER -= 1
        else:
            ret, img = webcam.read()
            cv2.imshow('a', img)
            cv2.waitKey(2000)
            cv2.imwrite('captured_images/les.jpg', img)
    elif k == 27:
        break
webcam.release()
cv2.destroyAllWindows()
```
